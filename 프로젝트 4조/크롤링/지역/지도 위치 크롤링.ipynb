```python
import urllib.request  # 원격 서버 파일 요청
from bs4 import BeautifulSoup  # html 파싱
import pandas as pd  # 데이터프레임 처리

url = 'https://droneportal.or.kr/subList/22000000153'
res = urllib.request.urlopen(url)  # 웹 페이지 요청
data = res.read()  # 페이지 데이터를 읽음
src = data.decode('utf-8')  # 데이터를 UTF-8로 디코딩

# BeautifulSoup으로 HTML 파싱
html = BeautifulSoup(src, 'html.parser')

# 모든 <td> 태그를 찾음
location = html.find_all('td')

# 'td' 태그 내의 내용을 리스트에 저장하고 None 값은 제외
locate = [loc.string.strip() for loc in location if loc.string]

# locate 리스트를 3개씩 묶어서 2차원 리스트로 변환
chunked_data = [locate[i:i+3] for i in range(0, len(locate), 3)]

# 데이터프레임 생성
locate = pd.DataFrame(chunked_data, columns=['Column 1', 'Column 2', 'Column 3'])
```
