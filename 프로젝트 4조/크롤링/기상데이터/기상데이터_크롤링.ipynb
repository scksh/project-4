## 2) 데이터 요청 및 추출

### 2.1) 데이터 요청&추출 함수 정의

```python
def fetch_weather_data(url, api_key, startDt, endDt):
    API_KEY_decode = requests.utils.unquote(api_key) # 디코딩
    dfs_lst = [] 
    for stnIds in range(90,296): # 지점 코드 전체 범위 반복
        try:
            print(f'종관기상관측 지점 {stnIds}번 데이터 추출:')
            # 요청 변수
            API_KEY_decode       # surviceKey
            pageNo = '1'         # 페이지 수
            numOfRows = '366'    # 한 페이지당 표출 데이터 수
            dataType = 'JSON'    # 요청자료형식(XML/JSON), Default: XML
            dataCd = 'ASOS'      # 자료 분류 코드(ASOS)
            dateCd = 'DAY'       # 날짜 분류 코드(DAY)
            startDt = startDt # 조회 기간 시작일(YYYYMMDD)
            endDt = endDt   # 조회 기간 종료일(YYYYMMDD) (전일 D-1 까지 제공)
            stnIds = stnIds       # 종관기상관측 지점 번호
    
            # 요청 변수 저장
            params ={'serviceKey' : API_KEY_decode,
                     'pageNo' : pageNo, 
                     'numOfRows' : numOfRows,
                     'dataType' : dataType, 
                     'dataCd' : dataCd,
                     'dateCd' : dateCd, 
                     'startDt' : startDt,
                     'endDt' : endDt, 
                     'stnIds' : stnIds }
    
            # 지점 번호(stnIds)에 따른 데이터 요청
            r = requests.get(url, params = params)
            # data to json
            dict_data = r.json()
            # json to DataFrame
            df = pd.DataFrame(dict_data['response']['body']['items']['item'])[['stnId', 'stnNm', 'tm', 'avgWs', 'avgTa', 'sumRn']]
            dfs_lst.append(df)
            print('성공')
            print()
        except:
            print(f'지점이 존재하지 않음')
            print()
            continue   
    df_weather = pd.concat(dfs_lst, ignore_index=True) # 모든 지점 데이터 concat
    return df_weather
```
